```python
from pandas_datareader import data as pdr
import yfinance as yf
import pandas as pd
import time
import json
import pdb
```

```python
# Run the following command to install the get_all_tickers module
# pip install get_all_tickers
from get_all_tickers import get_tickers as gt
list_of_tickers = gt.get_tickers()
```

```python
with open('ticker_industry.json') as f:
    json_data = json.load(f)
```

```python
import numpy as np
test = np.array(list(json_data.values()))
result = []
for lst in list(json_data.values()):
    result += lst

print(len(json_data.keys()))
len(result)
```

```python
t1 = time.time()
outliers = []
data = pd.DataFrame()
ticker_industry = json_data
ticker_outlier = []

data.loc['B','Country'] = yf.Ticker('B').info['country']
data = pd.DataFrame(yf.Ticker('B').history(start='2020-01-01',end='2020-04-30')['Close'].to_dict(),index=['B'])

for ticker in list_of_tickers:
#     if(ticker not in result):
#     pdb.set_trace()
#         print(ticker," not in json file")
        try:
            ticker_obj = yf.Ticker(ticker)
            data.loc[ticker_obj.ticker] = ticker_obj.history(start='2020-01-01',end='2020-04-30')['Close'].to_list()
            try:
                if(ticker_obj.info['industry'] in ticker_industry.keys()):
                    key = ticker_obj.info['industry']
                    ls = ticker_industry[key]
                    ticker_industry[key] = ls+[ticker]
                else:
                    ticker_industry[ticker_obj.info['industry']] = [ticker]  
                    data.loc[ticker,'Country'] = ticker_obj.info['country']
            except:
                ticker_outlier.append(ticker)
#                 continue
            try:
                data.loc[ticker,'Country'] = ticker_obj.info['country']
            except:
                print('Country not found for ',ticker)
        except:
            outliers.append(ticker)
```
