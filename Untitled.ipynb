```python
import yfinance as yf
import pandas as pd
import time

t1 = time.time()
outliers = []
data = pd.DataFrame()
df=pd.read_csv('generic.csv')

data = pd.DataFrame(yf.Ticker('B').history(start='2020-01-01',end='2020-04-30')['Close'].to_dict(),index=['B'])
for ticker in df['Ticker'].to_list():
    try:
        ticker_obj = yf.Ticker(ticker)
        print('Data found for ',ticker)
        data.loc[ticker_obj.ticker] = ticker_obj.history(start='2020-01-01',end='2020-04-30')['Close'].to_list()
        print('Added to data')
    except:
        outliers.append(ticker)
        print('No data found for ',ticker)
        
print(time.time()-t1)
```
