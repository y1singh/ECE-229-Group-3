```python
import pandas as pd
import numpy as np
```

```python
air_china_url = 'https://raw.githubusercontent.com/y1singh/ECE-229-Group-3/global_spread_analysis_chang/601111.SS.csv'
air_china = pd.read_csv(air_china_url, error_bad_lines=False)
air_china.head()
```

```python
air_china['Date'] = pd.to_datetime(air_china['Date']) 
mask = (air_china['Date'] >= '2020-01-22')
air_china=air_china.loc[mask]
```

```python
df1 = air_china[['Date','Close']].reset_index(drop=True).copy()
df1
```

```python
lst = [1]*82
lst = pd.Series(lst)
df1['Close'] = df1['Close']+lst
df1
```

```python
airlines = {'China Southern Airlines': '600029',
            'China Eastern Airlines': '600115',
            'Hainan Airlines': '600221',
            'Spring Airlines': '601021', # 'ji_xiang': '603885',
            'Guangzhou Baiyun International Airport': '600004',
            'Shanghai International Airport': '600009',
            'Xiamen International Airport': '600897',
            
            
}
git_base = 'https://raw.githubusercontent.com/y1singh/ECE-229-Group-3/global_spread_analysis_chang/'
```

```python
count=0
for code in list(airlines.values()):
    url = git_base + code + '.SS.csv'
    frame = pd.read_csv(url, error_bad_lines=False)
    frame['Date'] = pd.to_datetime(frame['Date']) 
    mask = (frame['Date'] >= '2020-01-22')
    frame=frame.loc[mask].reset_index(drop=True)
    df1['Close'] = df1['Close']+frame['Close']
    count+=1
```

```python
df1['Close'] = df1['Close'] / (len(airlines)+1)
```
